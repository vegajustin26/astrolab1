```python
import numpy as np
from scipy.special import factorial
import matplotlib.pyplot as plt

def binom_prob(k,n,p):
    prob1 = factorial(n)/(factorial(k)*factorial((n-k)))
    prob2 = p**k * (1-p)**(n-k)
    return prob1*prob2


n_trials = 9
p_success = 1.0/6.0
k_successes = np.linspace(0,n_trials,num=(n_trials+1))
p_array = binom_prob(k_successes,n_trials,p_success)

plt.figure()
ax = plt.subplot(111)
ax.plot(k_successes,p_array,'ks',ms=12)
ax.set_xlabel("k (number of successes)",fontsize="large")
ax.set_ylabel("Probability of k successes in {} trials".format(n_trials),fontsize="large")
ax.tick_params(labelsize="large")
ax.set_xlim(0,n_trials*1.1)
ax.set_ylim(0,max(p_array)*1.1)
ax.set_title("Rolling a 6-sided die {} times".format(n_trials))
ax.set_xticks(k_successes)
ax.grid()
```
